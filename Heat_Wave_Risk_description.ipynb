# Risk Assessment for Heat wave

### Risk assessment methodology

The risk assessment methodology is based on the combination of overheated areas and distribution of the vulnerable population, and areas with and usually overcrowded during the day. The identification of the overheated areas is based on the Land Surface Temperature product from the Landsat8 satellite imagery from the RSLAB webpage [https://rslab.gr/Landsat_LST.html]. The distribution of the vulnerable population, people over 65 years, is from the worldpop webpage [https://data.humdata.org/dataset/worldpop-age-and-sex-structures-for-slovakia]. The data about places that are regularly overcrowded are from the Zilina city office. 

Firstly, we downloaded the Landsat8 Land surface temperature (LST) for the summer months of June, July, and August. We created a raster stack from all LST images and calculated the one layer that represents the highest value for each pixel. We reclassify the raster into 5 classes based on the LST (<20, 20-30, 30-40, 40-50, 50< Celsius).

Secondly, we downloaded the world population data, for the group of the population and classified the raster into 5 equal interval classes.

Subsequently, we calculate the Heatwave risk layer by multiplying the LST and population layers, and we get a layer with values from 1 to 25, we also added the Risk matrix for risk assessment interpretation (Fig 2.).

Finally, we plotted the overcrowded places, which together with the Heat wave risk map highlighted the most important places (Fig 1.). 

From the Hazard assesment (Look to Heat-wave hazard assessment) we find out that the that the Heat-wave occurence trend should be more frequent in the future. However, we did not use these results directly in the Heat-wave risk map, because the resolution of the climate scenarios is too large for the small city like Zilina.
